```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split, GridSearchCV,StratifiedKFold
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,log_loss
from sklearn.svm import SVC
```

```python
bcancer = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Wisconsin/BreastCancer.csv",index_col=0)
```

```python
lbl = LabelEncoder()
X = bcancer.drop('Class',axis = 1)
y = pd.DataFrame(lbl.fit_transform(bcancer['Class']))
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=23,test_size=0.3,stratify=y)

svm = SVC(probability=True,random_state=23)
lr = LogisticRegression()
nb = GaussianNB()

voting = VotingClassifier(estimators=[("SVM",svm),("LR",lr),("NB",nb)],voting='soft')
voting.fit(X_train,y_train)

y_pred_prob = voting.predict_proba(X_test)[:,1]
print("Logloss : ",log_loss(y_test,y_pred_prob))
y_pred = voting.predict(X_test)
print("Accuraccy Score : ",accuracy_score(y_test,y_pred))
```

```python
kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
print(voting.get_params())
params = {'SVM__gamma':['scale','auto'],
          'SVM__C':np.linspace(0.001,5,5),
          'LR__penalty':['l1','l2','elastic',None],
          'NB__var_smoothing':np.linspace(0.0001,0.999,20)}
gcv = GridSearchCV(voting,param_grid=params,cv=kfold,scoring='neg_log_loss')

gcv.fit(X,y)
print("Best Parameters : ",gcv.best_params_)
print("Best Score : ",gcv.best_score_)
```

# Image Segmentation Dataset

```python
data = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Image Segmentation/Image_Segmention.csv")

lbl = LabelEncoder()
X = data.drop('Class',axis = 1)
y = pd.DataFrame(lbl.fit_transform(data['Class']))

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=23,test_size=0.3,stratify=y)

svm = SVC(probability=True,random_state=23)
lr = LogisticRegression()
nb = GaussianNB()

#model Assembling
voting = VotingClassifier(estimators=[("SVM",svm),("LR",lr),("NB",nb)],voting='soft')
voting.fit(X_train,y_train)

y_pred_prob = voting.predict_proba(X_test)#[:,1]
print("Logloss : ",log_loss(y_test,y_pred_prob))
y_pred = voting.predict(X_test)
print("Accuraccy Score : ",accuracy_score(y_test,y_pred))

#gridsearchcv
kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
print(voting.get_params())
params = {'SVM__gamma':['scale','auto'],
          'SVM__C':np.linspace(0.001,5,5),
          'LR__penalty':['l1','l2','elastic',None],
          "LR__multi_class":['ovr','multinomal'],
          'NB__var_smoothing':np.linspace(0.0001,0.999,20)}
gcv = GridSearchCV(voting,param_grid=params,cv=kfold,scoring='neg_log_loss')

gcv.fit(X,y)
print("Best Parameters : ",gcv.best_params_)
print("Best Score : ",gcv.best_score_)
```

# in voting Classifier there is a weight parameter that needs to be fine tune

---

**if we calculate the indvidual scores of each of the estimators(using gcv only), then the best performer should get best weight and worst performergets low weight**

**In Above Case LR has best score followed by SVM and NB shows poor performancwe so we assign weigths to lr as 0.5 , svm as 0.4 and nb as 0.1, Note these wieghts are completely random**


```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=23,test_size=0.3,stratify=y)

svm = SVC(probability=True,random_state=23)
lr = LogisticRegression()
nb = GaussianNB()

voting = VotingClassifier(estimators=[("SVM",svm),("LR",lr),("NB",nb)],voting='soft',weights=[0.5,0.4,0.1])
voting.fit(X_train,y_train)

y_pred_prob = voting.predict_proba(X_test)#[:,1]
print("Logloss : ",log_loss(y_test,y_pred_prob))
y_pred = voting.predict(X_test)
print("Accuraccy Score : ",accuracy_score(y_test,y_pred))
```

```python
#gridsearchcv
kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
print(voting.get_params())
params = {'SVM__gamma':['scale','auto'],
          'SVM__C':np.linspace(0.001,5,5),
          'LR__penalty':['l1','l2','elastic',None],
          "LR__multi_class":['ovr','multinomal'],
          'NB__var_smoothing':np.linspace(0.0001,0.999,20)}
gcv = GridSearchCV(voting,param_grid=params,cv=kfold,scoring='neg_log_loss')

gcv.fit(X,y)
print("Best Parameters : ",gcv.best_params_)
print("Best Score : ",gcv.best_score_)
```

from sklearn.ensemble import VotingRegressor

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import RandomizedSearchCV,train_test_split,KFold
from sklearn.ensemble import VotingRegressor
from sklearn.linear_model import ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.pipeline import Pipeline
```

```python
concrete = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Concrete Strength/Concrete_Data.csv")
```

```python
X = concrete.drop('Strength',axis = 1)
y = concrete['Strength']
```

```python
kfold = KFold(n_splits=5,shuffle=True,random_state=23)

elastic = ElasticNet()
std_scl = StandardScaler()
min_max = MinMaxScaler()
knn = KNeighborsRegressor()
dtr = DecisionTreeRegressor()

pipe = Pipeline([("SCL",std_scl),("KNN",knn)])

voting = VotingRegressor(estimators=[('ELASTIC',elastic),('PIPE',pipe),('DTR',dtr)])

params = {"ELASTIC__alpha":np.linspace(0.001,5,5),
          'ELASTIC__l1_ratio':np.linspace(0,1,5),
          'PIPE__KNN__n_neighbors':np.arange(1,10),
          'PIPE__SCL':[std_scl,min_max],
          'DTR__max_depth':[None,3,5],
          'DTR__min_samples_split':[2,5,10],
          'DTR__min_samples_leaf':[1,5,10]}


rgcv = RandomizedSearchCV(voting,param_distributions=params,cv=kfold,scoring='r2')

rgcv.fit(X,y)

print("Best Parameters : ",rgcv.best_params_)
print("Best Score : ",rgcv.best_score_)
```

```python
from sklearn.model_selection import GridSearchCV


rgcv = GridSearchCV(voting,param_grid=params,cv=kfold,scoring='r2')

rgcv.fit(X,y)

print("Best Parameters : ",rgcv.best_params_)
print("Best Score : ",rgcv.best_score_)
```

**Here we acn use Gridsearchcv but it takes too much time Here we havn't use weight parameter in votingregressor so to use that we have to check the individual performance of all the models**

**here R2 score for ElasticNet is 0.6006, for KNN-regressor it is 0.715 and for Decision tree regressor it is 0.847
so we are giving the more weightage to the DTR than KNN_R and less weightage to the ElasticNest**

**here we are taking weights equal to their corresponding R2 score
therefor we are using wights = [0.6006,0.715,0.847]**

```python
kfold = KFold(n_splits=5,shuffle=True,random_state=23)

elastic = ElasticNet()
std_scl = StandardScaler()
min_max = MinMaxScaler()
knn = KNeighborsRegressor()
dtr = DecisionTreeRegressor()

pipe = Pipeline([("SCL",std_scl),("KNN",knn)])

voting = VotingRegressor(estimators=[('ELASTIC',elastic),('PIPE',pipe),('DTR',dtr)],weights=[0.6006,0.715,0.847])

params = {"ELASTIC__alpha":np.linspace(0.001,5,5),
          'ELASTIC__l1_ratio':np.linspace(0,1,5),
          'PIPE__KNN__n_neighbors':np.arange(1,10),
          'PIPE__SCL':[std_scl,min_max],
          'DTR__max_depth':[None,3,5],
          'DTR__min_samples_split':[2,5,10],
          'DTR__min_samples_leaf':[1,5,10]}


rgcv = RandomizedSearchCV(voting,param_distributions=params,cv=kfold,scoring='r2')

rgcv.fit(X,y)

print("Best Parameters : ",rgcv.best_params_)
print("Best Score : ",rgcv.best_score_)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import RandomizedSearchCV,train_test_split,KFold
from sklearn.ensemble import VotingRegressor
from sklearn.linear_model import ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.pipeline import Pipeline
```

```python
data = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/CRAB_AGE/train.csv",index_col=0)
data1=pd.get_dummies(data,drop_first=True)
X= data1.drop("Age",axis=1)
y = data.Age
```

```python
kfold = KFold(n_splits=5,shuffle=True,random_state=23)

elastic = ElasticNet()

knn = KNeighborsRegressor()
dtr = DecisionTreeRegressor()



voting = VotingRegressor(estimators=[('ELASTIC',elastic),('KNN',knn),('DTR',dtr)])#,weights=[0.6006,0.715,0.847]

params = {"ELASTIC__alpha":np.linspace(0.001,5,5),
          'ELASTIC__l1_ratio':np.linspace(0,1,5),
          'KNN__n_neighbors':np.arange(1,10),
          'DTR__max_depth':[None,3,5],
          'DTR__min_samples_split':[2,5,10],
          'DTR__min_samples_leaf':[1,5,10]}


rgcv = RandomizedSearchCV(voting,param_distributions=params,cv=kfold,scoring='r2')

rgcv.fit(X,y)

print("Best Parameters : ",rgcv.best_params_)
print("Best Score : ",rgcv.best_score_)
```

```python
#inferencing
test_data = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/CRAB_AGE/test.csv",index_col=0)
test_data = pd.get_dummies(test_data,drop_first=True)
best_model = rgcv.best_estimator_

y_pred = best_model.predict(test_data)

submission={'id':np.arange(74051,123419),'Age':y_pred}
submission=pd.DataFrame(submission)
submission.to_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/CRAB_AGE/submission.csv",index=False)
```
