# ***Missing Value Imputation***

# ***1. Simple Imputer using Strategy***

```python
from sklearn.impute import SimpleImputer
import pandas as pd

chem  = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Chemical Process Data/ChemicalProcess.csv")
```

```python
chem.isnull().sum()
```

```python
# Here The Standard Imputer Strategy Is Mean i.e Null Values Are Getting Filled With Mean Of Remaining
imp = SimpleImputer()
imputed = imp.fit_transform(chem)
pd_imp = pd.DataFrame(imputed,columns=chem.columns)
pd_imp.isnull().sum()
```

# ***2. KNN Imputer***

```python
from sklearn.impute import KNNImputer
import numpy as np

x = np.array([[1,2,np.nan],
             [8,6.3,5.6],
             [1.2,2.3,4],
             [1.25,2.35,5],
             [1.2,2.4,3],
             [np.nan,6,5],
             [10,6.43,5.61]])

imputer = KNNImputer(n_neighbors=1)
imputer.fit_transform(x)
```

```python
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import ElasticNet
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold
from sklearn.impute import SimpleImputer,KNNImputer
import numpy as np

chem = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Chemical Process Data/ChemicalProcess.csv")

X = chem.drop('Yield',axis =1)
y = chem['Yield']

imp = SimpleImputer(strategy='mean') #stratergy=mean
elastic = ElasticNet()
pipe = Pipeline([('Impute',imp),('Elastic',elastic)])
kfold = KFold(n_splits=5,shuffle=True,random_state=23)
l1_rat = np.linspace(0.001,0.999,20)
alphas = np.linspace(0.001,4,20)
params = {'Elastic__alpha':alphas,'Elastic__l1_ratio':l1_rat,'Impute__strategy':['mean','median']}


gcv = GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='r2')
gcv.fit(X,y)
print("Best Parameters : ",gcv.best_params_)
print("Best Score : ",gcv.best_score_)
```

```python
knn_imp = KNNImputer()
pipe = Pipeline([('Impute',knn_imp),('Elastic',elastic)])
params = {'Elastic__alpha':alphas,'Elastic__l1_ratio':l1_rat,'Impute__n_neighbors':np.arange(1,6)}
gcv = GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='r2')
gcv.fit(X,y)
print("Best Parameters : ",gcv.best_params_)
print("Best Score : ",gcv.best_score_)
```
