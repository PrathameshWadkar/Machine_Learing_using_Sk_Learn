```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, log_loss
from sklearn.model_selection import train_test_split,KFold,StratifiedKFold,GridSearchCV
from sklearn.preprocessing import LabelEncoder

cancer = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Wisconsin/BreastCancer.csv")

lbl = LabelEncoder()
X = cancer.drop('Class',axis=1)
y=cancer['Class']
y = pd.DataFrame(lbl.fit_transform(y))

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,stratify=y,random_state=23)
```

#  ***SVR***

```python
from sklearn.svm import SVR

import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, log_loss
from sklearn.model_selection import train_test_split,KFold,StratifiedKFold,GridSearchCV
from sklearn.preprocessing import LabelEncoder,StandardScaler,MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold

concrete =  pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Concrete Strength/Concrete_Data.csv")

X = concrete.drop('Strength',axis = 1)
y = concrete['Strength']

kfold = KFold(n_splits=5,shuffle=True,random_state=23)


svr = SVR(kernel='linear')


scalar = StandardScaler()
scalar1 = MinMaxScaler()
pipe = Pipeline([("SCL",scalar),("svr",svr)])
params = {"svr__C":np.linspace(0.001,5,10),"svr__epsilon":np.linspace(0.001,5,10),"SCL":[scalar,scalar1]}

gcv = GridSearchCV(pipe,cv=kfold,param_grid=params)
gcv.fit(X,y)

print(gcv.best_params_)
print(gcv.best_score_)

###########################
     # Inferencing #
############################
test_data = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Concrete Strength/testConcrete.csv")

bm = gcv.best_estimator_
y_pred = bm.predict(test_data)
print(y_pred)
```

```python
y=concrete[['Strength']]
#using Standard Scaler
y_scaler=StandardScaler()
scl_y= y_scaler.fit_transform(y)

################ Linear ###########

pipe = Pipeline([("SCL",scalar),("svr",svr)])
params = {"svr__C":np.linspace(0.001,5,10),"svr__epsilon":np.linspace(0.001,5,10)}

gcv = GridSearchCV(pipe,cv=kfold,param_grid=params)
gcv.fit(X,scl_y[:,0])

print(gcv.best_params_)
print(gcv.best_score_)

############### RBF #################
svr = SVR(kernel='rbf')


scaler = StandardScaler()

pipe = Pipeline([("SCL",scaler),("svr",svr)])
params = {"svr__C":np.linspace(0.001,5,10),"svr__epsilon":np.linspace(0.001,5,10)}

gcv = GridSearchCV(pipe,cv=kfold,param_grid=params)
gcv.fit(X,scl_y[:,0])

print(gcv.best_params_)
print(gcv.best_score_)

###### Inferencing #############
test_data = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Concrete Strength/testConcrete.csv")

bm = gcv.best_estimator_
y_pred = bm.predict(test_data)
print(y_pred)

#since our data is scaled we are getting scled prediction
#now we  have to convert them into original form

min_y=np.min(scl_y)
max_y=np.max(scl_y)
y = y_pred*(max_y-min_y) + min_y
print(y)

#or

#y_scaler.inverse_transform(y_pred)
#using this we got folllowing error
#Reshape your data either using array.reshape(-1, 1) if your data has a single feature or array.
#reshape(1, -1) if it contains a single sample.
y_scaler.inverse_transform(y_pred.reshape(-1,1))
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,StandardScaler,MinMaxScaler
from sklearn.model_selection import RandomizedSearchCV,StratifiedKFold
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis,QuadraticDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
import warnings
warnings.filterwarnings(action='ignore')

#Loading the Data
data = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Satellite Imaging/Satellite.csv",sep=';')
```

```python
# initiating the LabelEncoder
lbl = LabelEncoder()

#splitng the data
X = data.drop("classes",axis=1)
y = pd.DataFrame(lbl.fit_transform(data.classes))
```

```python
#we are using Kfold Cross validation
kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
```

```python
#Initiating the model
#LDA
lda = LinearDiscriminantAnalysis()
#parameters tuning for LDA
params1 = {
    "solver": ['svd', 'lsqr', 'eigen']
}
#to get best result we are using RandomisedsearchCV
rgcv = RandomizedSearchCV(lda,param_distributions=params1,cv=kfold,scoring="neg_log_loss")
rgcv.fit(X,y)
print("Best parameters using LDA are :",rgcv.best_params_)
print("Best Score using LDA are :",rgcv.best_score_)
```

```python
#QDA
qda = QuadraticDiscriminantAnalysis()

#tunning the parameters
params={}

#to get best result we are using RandomisedsearchCV
rgcv = RandomizedSearchCV(qda,param_distributions=params,cv=kfold,scoring="neg_log_loss")
rgcv.fit(X,y)
print("Best parameters using QDA are :",rgcv.best_params_)
print("Best Score using QDA are :",rgcv.best_score_)
```

```python
#SVC
svm = SVC(kernel='linear',probability=True,random_state=23)
# Linear function With Scaling
#now to scaling the data using standard scaler
std_scl = StandardScaler()


#Creating the Pipeline to compute the result
pipe = Pipeline([('SCL',std_scl),('SVM',svm)])
params = {"SVM__decision_function_shape":['ovo'],
          "SVM__C":np.linspace(0.001,5,5)}

#to get best result we are using RandomisedsearchCV
rgcv = RandomizedSearchCV(pipe,param_distributions=params,cv=kfold,scoring="neg_log_loss")
rgcv.fit(X,y)
print("Best parameters using SVM(SVC:linear:standard_scaling) are :",rgcv.best_params_)
print("Best Score using SVM(SVC:linear:standard_scaling) are :",rgcv.best_score_)
```
