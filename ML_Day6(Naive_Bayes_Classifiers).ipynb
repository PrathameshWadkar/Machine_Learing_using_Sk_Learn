```python
import pandas as pd
from sklearn.naive_bayes import  BernoulliNB
import numpy as np
```

```python
tel = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Datasets/tel_bayes.csv")
dum_tel = pd.get_dummies(tel,drop_first=True)
#alpha can be any number including 0
# but if we take it as 0 it can produce infinity as (n/0) = infinity
#To avoid this we use Force_alpha = True which adds 1*e^-10 in alpha
#i.e (n/(0+1*e^+10))
nb = BernoulliNB(alpha=0,force_alpha=True )

X = dum_tel[['TT_gt_100_y','Gender_male']]
y = dum_tel['Response_not bought']

# Calculates Apriori Probabilty
nb.fit(X,y)
tst = np.array([["n","female"],
                ["n","male"],
                ["y","male"],
                ["y","female"]])
tst =pd.DataFrame(tst,columns = ['TT_gt_100','Gender'])
dum_tst = pd.get_dummies(tst,drop_first=True)

# Predict Proba Calculates Posterior Probabilty
nb.predict_proba(dum_tst)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.naive_bayes import BernoulliNB


telecom = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Datasets/Telecom.csv")
dum_telecom = pd.get_dummies(telecom,drop_first=True)
# print(dum_telecom)
X = dum_telecom.drop("Response_Y",axis=1)
y = dum_telecom['Response_Y']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=23,stratify=y)

nb = BernoulliNB()
nb.fit(X_train,y_train)#Model Buildin: Apriori Probs Calculated

y_pros = nb.predict_proba(X_test)#Postpriories Calculated
y_pred = nb.predict(X_test)# Applying Built test data

print(confusion_matrix(y_test,y_pred))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV,train_test_split,StratifiedKFold
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import BernoulliNB

cancer = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Cancer/Cancer.csv",index_col='subjid')

dum_cancer  = pd.get_dummies(cancer,drop_first=True)
X = dum_cancer.drop('Class_recurrence-events',axis = 1)
y = dum_cancer['Class_recurrence-events']

nb = BernoulliNB()

kfold = StratifiedKFold(n_splits=5,shuffle= True,random_state=23)
params = {'alpha':np.linspace(0,5,20)}
gcv =  GridSearchCV(nb,param_grid=params,cv=kfold,scoring='neg_log_loss')

gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

# ***Gaussian Naive Bayes Classifier***

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import  GridSearchCV,StratifiedKFold
```

```python
hr = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/human-resources-analytics/HR_comma_sep.csv")

dum_hr = pd.get_dummies(hr,drop_first=True)
X = dum_hr.drop('left',axis=1)
y = dum_hr['left']

#Performing K-fold Cross-Validation
kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=23)

#Instantiating the Model
gau = GaussianNB()

params = {'var_smoothing':np.arange(0,20)}
gcv = GridSearchCV(gau,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis,LinearDiscriminantAnalysis
from sklearn.model_selection import GridSearchCV,StratifiedKFold
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np
```

```python
brupt = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Company Bankruptcy/data.csv")
X = brupt.drop('Bankrupt?',axis = 1)
y = brupt['Bankrupt?']

########## LDA ##################
lda = LinearDiscriminantAnalysis()

kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
params = {'solver':['svd', 'lsqr', 'eigen']}

gcv = GridSearchCV(lda,param_grid=params,cv=kfold,scoring="neg_log_loss")
gcv.fit(X,y)
print("Best Score : ",gcv.best_score_)

######## QDA #############
qda = QuadraticDiscriminantAnalysis()
kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
params = {}
gcv = GridSearchCV(qda,param_grid=params,cv=kfold,scoring='neg_log_loss')

gcv.fit(X,y)
print("Best score : ",gcv.best_score_)
```
