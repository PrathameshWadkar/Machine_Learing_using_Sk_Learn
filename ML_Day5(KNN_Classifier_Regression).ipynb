# ***1. KNN Classification***

```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,log_loss
```

```python
data = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Datasets/RidingMowers.csv")
```

```python
lbl = LabelEncoder()
data['Response']=lbl.fit_transform(data.Response)

X = data.drop('Response',axis =1)
y = data.Response

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 23,stratify = y)

#Instantiating the KNN model
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)

#Calculating the the accuracy
print(accuracy_score(y_test,y_pred))

#calculating the probability Loss
y_pred_prob = knn.predict_proba(X_test)
print(log_loss(y_test,y_pred_prob))
```

```python
# For GridSearchCV

from sklearn.model_selection import GridSearchCV,KFold

kfold = KFold(n_splits=5,shuffle=True,random_state=23)
knn = KNeighborsClassifier()

params = {"n_neighbors":np.arange(1,11)}

#logloss
gcv = GridSearchCV(knn,param_grid=params,cv=kfold,scoring='neg_log_loss')

gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,log_loss

#loading the data
data = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Image Segmentation/Image_Segmention.csv")

#Label Encoding
lbl = LabelEncoder()
data['Class']=lbl.fit_transform(data.Class)

#Creating DataFreame
X = data.drop('Class',axis =1)
y = data.Class

#Spliting the data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 23,stratify = y)

#Instantiating the KNN model
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)

#Calculating the the accuracy
print("Accuracy Score :",accuracy_score(y_test,y_pred))

#calculating the probability Loss
y_pred_prob = knn.predict_proba(X_test)
print("Log_loss :",log_loss(y_test,y_pred_prob))

#finding Best Parameters
from sklearn.model_selection import GridSearchCV,KFold

kfold = KFold(n_splits=5,shuffle=True,random_state=23)
#tunning the N-neighbor Parameter
knn = KNeighborsClassifier()

params = {"n_neighbors":np.arange(1,11)}

#logloss
gcv = GridSearchCV(knn,param_grid=params,cv=kfold,scoring='neg_log_loss')

gcv.fit(X,y)
print("Best Parameters : ",gcv.best_params_)
print("Best score : ",gcv.best_score_)
```

# ***2. KNN Regression***

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,log_loss
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.pipeline import Pipeline

boston  = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Datasets/Boston.csv")

X = boston.drop('medv',axis = 1)
y = boston['medv']

from sklearn.neighbors import KNeighborsRegressor

# kfold = KFold(n_splits=5,shuffle=True,random_state=23)
# knn = KNeighborsRegressor()
# params = {'n_neighbors' : np.arange(1,11)}
# gcv=GridSearchCV(knn,param_grid=params,cv=kfold,scoring='r2')
# gcv.fit(X,y)
# print(gcv.best_params_)
# print(gcv.best_score_)


#finding Best Parameters
from sklearn.model_selection import GridSearchCV,KFold

kfold = KFold(n_splits=5,shuffle=True,random_state=23)
scaler = StandardScaler()
mn = MinMaxScaler()
knn = KNeighborsRegressor()


#Instantiating the Pipeline Object
pipe = Pipeline([("SCL",scaler),('KNN',knn)])
params = {"KNN__n_neighbors":np.arange(1,11),
          "SCL":[scaler,mn]}

#logloss
gcv = GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='r2')

gcv.fit(X,y)
print("Best Parameters : ",gcv.best_params_)
print("Best score : ",gcv.best_score_)
```

# Trying On Insurance Dataset

---
For comparing KNN-Regressor and ElasticNet


```python
import pandas as pd
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV,KFold
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.pipeline import Pipeline

data = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Medical Cost Personal/insurance.csv")

#Since the Data is not in Interger format we have to use get_dummies
data = pd.get_dummies(data,drop_first=True)

X = data.drop('charges',axis = 1)
y = data['charges']

#Instantiating all Models
knn = KNeighborsRegressor()
scaler = StandardScaler()
mn = MinMaxScaler()
elastic = ElasticNet()

#using K-fold Cross validation
kfold = KFold(n_splits=5,shuffle = True,random_state=23)



############# KNN-Regresssor Model ###########

#creating the Pipeline for KNN-Regressor
pipe1 = Pipeline([("SCL",scaler),('KNN',knn)])

#Declaring the parameters
params = {'SCL': [scaler,mn],
          'KNN__n_neighbors':np.arange(1,11)}

#Geeting the Best Result
gcv = GridSearchCV(pipe1,param_grid=params,cv = kfold,scoring='r2')
gcv.fit(X,y)
print("KNN-Regerssor Best Parameters : ",gcv.best_params_)
print("KNN-Regressor Best Score : ",gcv.best_score_)




################ElasticNet Model###################

#Creating Pipeline for Elasticnet
pipe2 = Pipeline([("SCL",scaler),('ELS',elastic)])

#Decalring the Alpha and L1_ratio
params = {'SCL': [scaler,mn],
          'ELS__l1_ratio':np.linspace(0.001,0.999,20),
          'ELS__alpha':np.linspace(0.001,4,20)}

#Getting the Best Result
gcv = GridSearchCV(pipe2,param_grid=params,cv = kfold,scoring='r2')
gcv.fit(X,y)
print("ElasticNet Best Parameters : ",gcv.best_params_)
print("ElasticNet Best Score : ",gcv.best_score_)
```

# ***Comparing Logistic Regression Classifier and KNN-Classifier***

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,StandardScaler,MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline


data = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Vehicle Silhouettes-20231103T125438Z-001/Vehicle Silhouettes/Vehicle.csv")

lbl = LabelEncoder()
X= data.drop('Class',axis=1)
y=lbl.fit_transform(data.Class)

kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=23)

knn = KNeighborsClassifier()
sc = StandardScaler()
mn = MinMaxScaler()

pipe1 = Pipeline([("SCL",scaler),('KNN',knn)])
params = {"SCL":[sc,mn],
          "KNN__n_neighbors":np.arange(1,11)}

gcv = GridSearchCV(pipe1,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)
print("Best Parameters For KNN_Classifier",gcv.best_params_)
print("Best Score For KNN_Classifier",gcv.best_score_)



```

```python

lr = LogisticRegression()

pipe2 = Pipeline([("SCL",scaler),('LR',lr)])
params = {"SCL":[sc,mn],
          "LR__penalty":['l1','l2','ElasticNet',None],
        'LR__solver':['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga'],
        'LR__multi_class':['ovr','multinomial']}

gcv = GridSearchCV(pipe2,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)
print("Best Parameters For KNN_Classifier",gcv.best_params_)
print("Best Score For KNN_Classifier",gcv.best_score_)
```
