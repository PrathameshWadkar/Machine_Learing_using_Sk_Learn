# Supervised Learning Algorithm
# 1. Linear Regression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
boston = pd.read_csv('Boston.csv')
```

```python
X=boston.drop('medv',axis=1)
y=boston['medv']
```

```python
#intiating the models

lr=LinearRegression()
dtr=DecisionTreeRegressor()
```

```python
#spliting of the dataset
train,test = train_test_split(boston , test_size=0.3)
```

```python
X_train = train.drop('medv',axis=1)
y_train = train['medv']
X_test = test.drop('medv',axis=1)
y_test =test['medv']
```

```python
#fiting data into the Linear Regression model
lr.fit(X_train,y_train)

#coefficients
print("Coefficients :",lr.coef_)

#predicting on test data
y_pred =lr.predict(X_test)

#calculating
print("Mean Squared Error :",mean_squared_error(y_test,y_pred))
```

```python
#fiting data into the DecisionTreeRegressor model
dtr.fit(X_train,y_train)

#predicting on test data
y_pred =dtr.predict(X_test)

#calculating
print("Mean Squared Error :",mean_squared_error(y_test,y_pred))
```

```python
lr.fit(X_train,y_train)
dtr.fit(X_train,y_train)

y_pred = lr.predict(X_test)
print("mean_squared_error for Linear Regressor :",mean_squared_error(y_test,y_pred))


print("="*50)
y_pred = dtr.predict(X_test)
print("mean_squared_error for Decision tree Regressor :",mean_squared_error(y_test,y_pred))
```
