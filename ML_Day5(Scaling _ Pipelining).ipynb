# ***1. Standard Scaler***

```python
import numpy as np
X_train = np.array([[24000,8],
                   [30000,2],
                   [35000,7],
                   [34500,1]])

X_test = np.array([[34000,7],[12000,1]])

#Standard Scaling
col1 = X_train[:,0]
trf_col1= (col1 - col1.mean())/np.std(col1)
print(trf_col1.mean(),np.std(trf_col1))

col2 = X_train[:,1]
trf_col2= (col2 - col2.mean())/np.std(col2)
print(trf_col2.mean(),np.std(trf_col2))

# all Columns at Onces

means = X_train.mean(axis=0)
stds = X_train.std(axis=0)
print("On Train data",(X_train-means)/stds)
print("On test data",(X_train-means)/stds)

print("means",means)
print("Standared Deviation",stds)

#usnig library

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
print("Using the Library")
print(sc.fit_transform(X_train))
print(sc.transform(X_test))
print(sc.mean_)
print(sc.scale_)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,log_loss
from sklearn.preprocessing import StandardScaler

#loading the data
data = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Wisconsin/BreastCancer.csv")

#Label Encoding
lbl = LabelEncoder()
data['Class']=lbl.fit_transform(data.Class)

#Creating DataFreame
X = data.drop('Class',axis =1)
y = data.Class

#Spliting the data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 23,stratify = y)

#Transforming the data
sc = StandardScaler()
X_scl_train = sc.fit_transform(X_train)
X_scl_test = sc.fit_transform(X_test)

#Instantiating the KNN model
knn = KNeighborsClassifier(n_neighbors=1)

#fiting the model using Unscaled Data
knn.fit(X_scl_train,y_train)
y_pred = knn.predict(X_test)

#Calculating the the accuracy
print(accuracy_score(y_test,y_pred))

#calculating the probability Loss
y_pred_prob = knn.predict_proba(X_test)
print(log_loss(y_test,y_pred_prob))

########################################

#fiting the model using the Scaled data
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_scl_train,y_train)
y_scl_pred = knn.predict(X_scl_test)

#Calculating the the accuracy
print(accuracy_score(y_test,y_scl_pred))

#calculating the probability Loss
y_scl_pred_prob = knn.predict_proba(X_scl_test)
print(log_loss(y_test,y_scl_pred_prob))
```

# ***Pipelining***

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,log_loss
from sklearn.pipeline import Pipeline

scaler = StandardScaler()
knn = KNeighborsClassifier(n_neighbors=1)

#Instantiating the Pipeline Object
pipe = Pipeline([("SCL",scaler),('KNN',knn)])

#Fitting the Data
pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)
print(accuracy_score(y_test,y_pred))
y_pred_prob = pipe.predict_proba(X_test)
print(log_loss(y_test,y_pred_prob))
```

```python
#finding Best Parameters
from sklearn.model_selection import GridSearchCV,KFold

kfold = KFold(n_splits=5,shuffle=True,random_state=23)
scaler = StandardScaler()
knn = KNeighborsClassifier()
#Instantiating the Pipeline Object
pipe = Pipeline([("SCL",scaler),('KNN',knn)])
params = {"KNN__n_neighbors":np.arange(1,11)}

#logloss
gcv = GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='neg_log_loss')

gcv.fit(X,y)
print("Best Parameters : ",gcv.best_params_)
print("Best score : ",gcv.best_score_)
```

# BreastCancer Dataset

```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,log_loss
from sklearn.preprocessing import StandardScaler

#loading the data
data = pd.read_csv("/content/drive/MyDrive/KP_Module/Practical Machine Learning/Practise/Cases/Wisconsin/BreastCancer.csv")

#Label Encoding
lbl = LabelEncoder()
data['Class']=lbl.fit_transform(data.Class)

#Creating DataFreame
X = data.drop('Class',axis =1)
y = data.Class

#Spliting the data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 23,stratify = y)

################ Standard Scaler ########################

#Transforming the data
sc = StandardScaler()
X_scl_train = sc.fit_transform(X_train)
X_scl_test = sc.fit_transform(X_test)

#fiting the model using the Scaled data
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_scl_train,y_train)
y_scl_pred = knn.predict(X_scl_test)

#Calculating the the accuracy
print("Standard Scaling Accuracy",accuracy_score(y_test,y_scl_pred))

#calculating the probability Loss
y_scl_pred_prob = knn.predict_proba(X_scl_test)
print("Standard Scaling Log_Loss",log_loss(y_test,y_scl_pred_prob))



################# MinMax Scaler ########################

#Transforming the data using MinMax Scaler
mn = MinMaxScaler()
X_scl_train = mn.fit_transform(X_train)
X_scl_test = mn.fit_transform(X_test)

#fiting the model using the Scaled data
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_scl_train,y_train)
y_scl_pred = knn.predict(X_scl_test)

#Calculating the the accuracy
print("MinMax Scaling Accuracy :",accuracy_score(y_test,y_scl_pred))

#calculating the probability Loss
y_scl_pred_prob = knn.predict_proba(X_scl_test)
print("MinMax Scaling LogLoss",log_loss(y_test,y_scl_pred_prob))
```

```python
#finding Best Parameters
from sklearn.model_selection import GridSearchCV,KFold

kfold = KFold(n_splits=5,shuffle=True,random_state=23)
scaler = StandardScaler()
mn = MinMaxScaler()
knn = KNeighborsClassifier()


#Instantiating the Pipeline Object with Standard Scaler
pipe = Pipeline([("SCL",scaler),('KNN',knn)])
params = {"KNN__n_neighbors":np.arange(1,11),
          "SCL":[scaler,mn]}

#logloss
gcv = GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='neg_log_loss')

gcv.fit(X,y)
print("Best Parameters : ",gcv.best_params_)
print("Best score : ",gcv.best_score_)
```
