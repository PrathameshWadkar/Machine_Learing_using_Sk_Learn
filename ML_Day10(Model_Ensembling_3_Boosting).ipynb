# ***2. Gradient Boosting***
----

***A. GradienBoostClassifer***



```
from sklearn.ensemble import GradientBoostClassifier
```



````python
from sklearn.ensemble import GradientBoostingClassifier
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import KFold,GridSearchCV,train_test_split,KFold
from sklearn.metrics import log_loss,accuracy_score
from sklearn.ensemble import  RandomForestClassifier
import matplotlib.pyplot as plt
````

````python
cancer =pd.read_csv("Cases/Wisconsin/BreastCancer.csv")
lbl = LabelEncoder()

X = cancer.drop("Class",axis = 1)
y = pd.DataFrame(lbl.fit_transform(cancer['Class']))

gbc = GradientBoostingClassifier(random_state=23)

params = {'max_depth':[2,3,4,5,6,None],
          'learning_rate':np.linspace(0.001,0.999,10),
          'n_estimators':[10,50,150]}
kfold=KFold(n_splits=5,shuffle=True,random_state=23)
gcv = GridSearchCV(gbc,param_grid=params,cv =kfold,scoring='neg_log_loss')
gcv.fit(X,y)
print("best Parameters :",gcv.best_params_)
print("best Score :",gcv.best_score_)
````

````python
bm_gbc = gcv.best_estimator_

df_imp = pd.DataFrame({'variable':bm_gbc.feature_names_in_,
                       'importance':bm_gbc.feature_importances_})
df_imp = df_imp[df_imp['importance']>0.0001]
df_imp = df_imp.sort_values(by='importance')
plt.barh(df_imp['variable'],df_imp['importance'])
plt.ylabel('Importance')
plt.show()
````

````python
cancer =pd.read_csv("Cases/Wisconsin/BreastCancer.csv")
lbl = LabelEncoder()

X = cancer.drop("Class",axis = 1)
y = pd.DataFrame(lbl.fit_transform(cancer['Class']))

xgb = XGBClassifier(random_state=23)

params = {'max_depth':[2,3,4,5,6,None],
          'learning_rate':np.linspace(0.001,0.999,10),
          'n_estimators':[10,50,150]}
kfold=KFold(n_splits=5,shuffle=True,random_state=23)
gcv = GridSearchCV(xgb,param_grid=params,cv =kfold,scoring='neg_log_loss')
gcv.fit(X,y)
print("best Parameters :",gcv.best_params_)
print("best Score :",gcv.best_score_)
````
